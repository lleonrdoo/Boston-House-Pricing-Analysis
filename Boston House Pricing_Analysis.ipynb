## 2. Mempersiapkan dataset 

```python
dataset = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
dataset
```

```python
# Disini akan ditambahkan kolom 'Price' dari data target yang sebelumnya digunakan untuk mengukur harga rumah
dataset['Price'] = boston.target
```

```python
dataset.info()
```

```python
# Selanjutnya, akan dilihat deskripsi statistik dari dataset tersebut
dataset.describe()
```

```python
# Setelah dilihat, kita juga perlu mengecek apakah terdapat nilai yang kosong atau tidak
dataset.isnull().sum()
```

Dapat dilihat disini bahwa tidak ada data yang kosong ('0')

# 

## 3. Exploratory data analysis dan mengecek korelasi

```python
dtc = dataset.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(dtc, annot=True, cmap='coolwarm', center=0)
plt.title('Heatmap Korelasi')
plt.show()
```

```python
# Misalkan kita ingin melihat hubungan antara 'CRIM' dengan 'Price', maka dapat digunakan scatterplot berikut

plt.figure(figsize=(10, 6))
plt.scatter(dataset['CRIM'], dataset['Price'], s=100, alpha=0.6, edgecolors='w', linewidth=0.5)
plt.title('Scatter Plot of CRIM vs Price')
plt.xlabel('Crime Rate')
plt.ylabel('Price')
plt.show()
```

```python
# Selanjutnya, kita juga bisa melihat hubungan-hubungan data-data lainnya dengan data 'Price'

plt.figure(figsize=(10, 6))
plt.scatter(dataset['RM'], dataset['Price'], s=100, alpha=0.6, edgecolors='w', linewidth=0.5)
plt.title('Scatter Plot of RM vs Price')
plt.xlabel('Room')
plt.ylabel('Price')
plt.show()
```

```python
# Lalu, jika kita ingin melihat gambaran kasar dari regresi linear hubungan kedua data diatas, dapat digunakan fungsi 'reg'
sns.regplot(x="RM", y="Price", data=dataset)
```

```python
# Saya juga ingin melihat regresi linear dari hubungan 'Price' dengan 'LSTA'
sns.regplot(x="LSTAT", y="Price", data=dataset)
```

Dapat dilihat bahwa ini adalah bentuk korelasi yang negatif 

```python
# Namun, jika data yang berkorelasi hanya memiliki nilai biner (1 atau 0) maka regresi linearnya akan berbentuk sebagai berikut 
sns.regplot(x="CHAS", y="Price", data=dataset)
```

```python
# Dapat juga dilihat bahwa terdapat relasi invers dari dataset tersebut
# Relasi tersebut adalah relasi antara 'Price' dengan 'PTRATIO'

sns.regplot(x="PTRATIO", y="Price", data=dataset)
```

Masih banyak lagi korelasi yang dapat dilihat disini, namun saya akan lanjut ke bagian pembentukkan model
